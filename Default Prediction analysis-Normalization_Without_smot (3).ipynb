# Use Chi-Square test for feature selection for categorical variables

```python
# create a dataframe of categorical variables
cont1=data[['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2','PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']]
cont=['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2','PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
```

```python
# import library to perform chi-square test
from scipy.stats import chi2_contingency
import numpy as np
from sklearn.feature_selection import SelectPercentile, chi2
ALPHA=0.05 
```

```python
cont1 = cont1.replace(-1, 10)
cont1 = cont1.replace(-2, 11)
```

```python
# a dataframe for the chi-square test results
chi2_res = chi2(cont1[cont],data['default payment next month'])
df_chi2 = pd.DataFrame({'target': 'default payment next month','cat_feature': cont1[cont].columns,'chi2': chi2_res[0],'p-value': chi2_res[1],'rounded p': np.around(chi2_res[1], 3),'alpha': ALPHA})
df_chi2['H0'] = df_chi2['p-value'].apply(lambda x: 'Rejected' if x <= ALPHA else 'Fail to reject')
df_chi2['relation'] = df_chi2['H0'].apply(lambda x: 'Dependent' if x=='Rejected' else 'Independent')
df_chi2.sort_values(by='chi2', ascending=False)
```

from above we can see that only MARRIAGE feacture is independent with default payment ,When I drop MARRIAGE there is nothing change in classification metrices 
so I have include it in model

```python
# Calculate Vif is function to calculate Vif and gives output in form of no multicollinearity Columns
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
def calculate_vif(X):
    thresh = 10
    output = pd.DataFrame()
    k = X.shape[1]
    dropped_columns = []
    
    vif = [variance_inflation_factor(X.values, j) for j in range(X.shape[1])]
    
    for i in range(1,k):
        print("==>Iteration no.", i)
        print(vif)
        
        # Returns indices of the max element of the array in a particular axis.
        a = np.argmax(vif)
        
        print("Max VIF is for variable no.:", a)
        
        if vif[a] <= thresh :
            break
            
        if i == 1 :          
            output = X.drop(X.columns[a], axis = 1)
            vif = [variance_inflation_factor(output.values, j) for j in range(output.shape[1])]
            print(output.columns, X.columns)
        elif i > 1 :
            output.drop(output.columns[a], axis = 1, inplace = True)
            vif = [variance_inflation_factor(output.values, j) for j in range(output.shape[1])]
            print(output.columns)
            
        dropped_columns.append(X.columns[a]) 
        
        print("dropping columns",dropped_columns)
        print("output columns", output.columns)
        print('X-columns', X.columns)
        
        print("#######################\n#######################")
    return(output, dropped_columns)
```
